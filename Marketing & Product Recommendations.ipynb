# Recommendations, Analysis & Insights

There are 7725 one-time buyers and 8406 repeat buyers. This rounds up to a retention rate of 52%. Revised cohort analysis based on CLV modeling revealed two different ways that we can segment the customers: 1) One-time vs Repeat Customers and 2) Personal vs Catering Orders. 

1.	Retention 

1)	Investigate the failure of Call Center. Call center was paused after a one month pilot, but need to dive deep into the actual methodology, logistics, cost, and even customer verbatim feedback or Net Promoter Score to understand its failure in customer retention. After all, Call is a decently effective method for acquiring new customers and for repeat customers.  

2)	Make Android users stick. Android is a newly rising star for onetime buyers contributing significantly for the increase in order size in 2019, but iOS is a more preferred method for repeat buyers. A demographic analysis of Android vs ios app users will help understand what inherent trait about the android users prevent them from making another purchase. Perhaps Android users prefer a vertical nav bar as opposed to horizontal, or they prefer to see fewer “sort by” categories so that they can spend less time searching. Maybe they care less about the “lower fee” options and want to find restaurants that offer better reward programs for 2nd, 3rd purchases. A deep dive cohort analysis of the ios vs android users will help us optimize the app design and promotion strategies accordingly as apps are the key growth areas highlighted in my analyses. 

4)	Prompt iOS users to keep making large orders. The cohort that makes less frequent but big orders will most likely be business orders, for social/networking events that are held on a regular basis. Come up with an in-app option for group orders. Allow team members to each write their desired items and let restaurant/order options be searchable by per person budget size to make the event organizer’s ordering task easier. Providing this level of personalized convenience will prompt users to go digital and make less phone calls.  

5)	Hone in on Weekdays, especially Tuesdays and Thursdays. Weekend lunches are the most popular, but Tuesday and Thursday orders have shown to provide greater CLV. It may be the companies hosting meetups/happy hours on Thursdays that we can, or pizza lovers who can consume pizza like bagels for weekday lunch or dinners that we can newly acquire. 

6)	Investigate the fluctuation in order size. The local dips (July, October, January) coincide with the dips in majority of sources, so this maybe due to do seasonal promotions/rewards or how the media budget was allocated throughout the quarter. We need more data for a closer look. 


2.	Acquisition 

1)	Give promotions and discounts for first purchases. Onetime buyers spend more money on average per transaction that repeat customers. While this may be due to net new users preferring to use the platform for making larger orders, it may also be that they’re simply not receiving reward/promotions earned from previous purchases. Cheaper first purchase often results in customer satisfaction and less likelihood to delete the app. 

2)	Conduct user research, a set of qualitative interviews or surveys across both consumers and pizzerias. Do people still prefer placing orders the old-fashioned way when it comes to pizza vs other food? Or can the company digitize all orders so that Web, Google, apps are the only sources and contribute to digitally transforming the customers’ purchasing behavior for greater cost efficiency? After all, the MVPs from the CLV-based cohort analysis strictly preferred Digital methods. 

3)	Grow late night orders. Dinner orders, while frequent, are mostly one-time purchases. Late night is the least frequent but a better contributor for CLV. How can we make spontaneous late night pizza runs digital? Do customers feel that ordering digitally kill the spontaneity and thrill? Is it because they would rather order by slice than by pie at night? Should pickup be highlighted more than delivery in the app from 11pm to 5am? Or is it more of an awareness issue, that consumers don’t know the existence of a personalized app made exclusively for pizza? Surely every NYC resident has had a 99 cent pizza at an ungodly hour while shuddering in cold or hiding from rain. The late night market is big enough, has existed for decades and isn’t going anywhere any time soon, and now it’s about investigating ways to digitally tap into it. 

3.	Personal vs Business (Catering) Orders 

Investigating the top 10 customers and re-running K-means clustering with the CLV value added to each customer showed that the MVPs fall into either of the two categories: 1) those who make less frequent but larger purchases and 2) those who make frequent but smaller purchases. I hypothesized to the two groups to be business catering orders vs personal orders. 

Following are the 3 groups I got from K-means:  

1)	First group (Cluster 0): 

- Least volume (only 192 orders fall into this group), yet highest CLV ($161 per transaction)

- Prefer weekdays: Thursday, Tuesday, Friday, Wednesday

- Prefer Dinner & Morning

- Use Web, Google, iOS, Android 

2)	Second group (Cluster 1): 

- Most volume, lowest CLV 

- Includes call center orders (aka onetime buyers)

- No strong preference over day/time bracket/source 

3)	Third group (Cluster 2): 

- Very similar to Cluster 1, but slightly higher CLV (by 40 cents) 

- Prefers late night orders

- Comparatively prefers app orders 

Cluster 0 can be considered as the bulk of the business catering category, with least volume but highest CLV. These MVP customers prefer Morning and weekday orders, prefer weekdays, and digital methods (web/google/apps). 

Cluster 2 can be regarded as the personal order category. Both Cluster 1 and 2 are a mix of onetimers & repeat and span across different sources, but Late Night & App orders seem to contribute to a slightly higher CLV for Cluster 2. Cluster 2 is the "growth" segment, a cohort different from Cluster 0, but for whom the company can run targeted strategies to convert to truer MVPs. 

4.	Key Drivers 

I ran a regression analysis to understand the profitability drivers -- to see which buttons to press to increase an average customer's CLV.  Results show that it's better to target customers who order frequently than those who've made an order recently - as expected. Web and Google are the top sources to acquire MVP customers from. Thursday & Tuesdays are the top days and late night is the time bracket for which marketing tactics can focus on prompting orders during. 

Key things to avoid to enhance retention (things that resulted in one time purchase) are Call Center, Pickup, Cash, Monday, Sunday, and dinner orders. Pick up/Cash may have resulted in inconvenient/unintuitive customer experiences. Dinner is usually a popular time for dine-ins, so restaurants may not have been able to pay as much attention to orders made through the platform.

5.	Other things to note 

1)	Make people pay with credit card. Common knowledge at this point that storing the CC number in the app is convenient for the user hence higher likelihood of repurchasing. 

2)	Delivery has higher CLV, but keep investigating tactics to emphasize pick up vs delivery for personal vs business orders and by day/time. 

3)	Keep up the good work for Google & Web. They’re the top 2 key drivers for both retention & acquisition. Google's order size has been soaring since 2019 for repeat customers, perhaps due to Google retargeting tactics becoming refined over time with the accumulation of marketing interaction data.

```python
import lifetimes
from lifetimes.plotting import *
from lifetimes.utils import *
from lifetimes.fitters import *

# create a Frequency/Recency matrix with "Age" column added
# age = duration between a customer’s first purchase and the last date in the data set (in weeks)
data=data[['user_id', 'purchase_date', 'restaurant_total']]
data['purchase_date'] = pd.to_datetime(data['purchase_date'],yearfirst=True)
data = summary_data_from_transaction_data(data, 'user_id', 'purchase_date', freq='W')
data.head()

# frequency distribution
data ['frequency'].plot(kind='hist', bins=50)
plt.title('Frequency Distribution')
print(data['frequency'].describe())
print(sum(data ['frequency'] == 0)/float(len(data)))

# recency distribution
data ['recency'].plot(kind='hist', bins=50)
plt.title('Recency Distribution')
print(data['recency'].describe())
print(sum(data ['recency'] == 0)/float(len(data)))

# BD/BG Fitter
from lifetimes import BetaGeoFitter
bgf = BetaGeoFitter(penalizer_coef=0.01)
bgf.fit(data['frequency'], 
        data['recency'], 
        data['T'],
       verbose=True)
print(bgf.params_)

#MBD/BG Fitter
from lifetimes import ModifiedBetaGeoFitter
mbg = ModifiedBetaGeoFitter(penalizer_coef=0.01)
mbg.fit(data['frequency'], 
        data['recency'], 
        data['T'],
       verbose=True)
print(mbg.params_)

# hold out period 
transaction_data = pd.read_csv('CLTV_BeforeLTV.csv')
from lifetimes.utils import calibration_and_holdout_data
summary_cal_holdout = calibration_and_holdout_data(transaction_data, 'user_id', 'purchase_date',
                                        calibration_period_end='2018-10-15', observation_period_end='2019-02-28' ) 
summary_cal_holdout.head()

from lifetimes.plotting import plot_period_transactions
plot_period_transactions(bgf)
plt.title('BD/BG: Frequency of Repeat Transactions')
plot_period_transactions(mbg)
plt.title('MBD/BG: Frequency of Repeat Transactions')

from lifetimes.plotting import plot_calibration_purchases_vs_holdout_purchases
plot_calibration_purchases_vs_holdout_purchases(bgf, summary_cal_holdout)
plt.title('Actual Purchases in Holdout Period vs BD/BG Predicted Purchases')
plot_calibration_purchases_vs_holdout_purchases(mbg, summary_cal_holdout)
plt.title('Actual Purchases in Holdout Period vs MBD/BG Predicted Purchases')
```

```python
# GGF only looks at repeat customers: so only selecting customers with at least one repeat purchase

data = pd.read_csv('CLTV_BeforeLTV.csv')
data=data[['user_id', 'purchase_date', 'restaurant_total']]
data['purchase_date'] = pd.to_datetime(data['purchase_date'],yearfirst=True)
data = summary_data_from_transaction_data(data, 'user_id', 'purchase_date', monetary_value_col='restaurant_total')

returning_customers_summary = data[data['frequency']>0]
returning_customers_summary.head()

# This model assumes that there is no relationship between the monetary value and the purchase frequency. Luckily the correlation is -0.037, low enough to use the GGF model. 
returning_customers_summary[['monetary_value', 'frequency']].corr()

from lifetimes import GammaGammaFitter
ggf = GammaGammaFitter(penalizer_coef = 0)
ggf.fit(returning_customers_summary['frequency'],
        returning_customers_summary['monetary_value'])
print(ggf)
p,q,v = ggf._unload_params('p', 'q', 'v')

print(ggf.conditional_expected_average_profit(
        data['frequency'],
        data['monetary_value']
    ).head(15))

print("Expected conditional average profit: %s, Average profit: %s" % (
ggf.conditional_expected_average_profit(
data['frequency'],
data['monetary_value']
).mean(),
data[data['frequency']>0]['monetary_value'].mean()
))

# top 10 predicted purchases in the next 3 months according to mbg
t=90
data['predicted_purchases_mbg'] = mbg.conditional_expected_number_of_purchases_up_to_time(t, data['frequency'], data['recency'], data['T'])
best_projected_cust_mbg = data.sort_values(by='predicted_purchases_mbg').tail(10)
best_projected_cust_mbg

# let's map out expected average profit and total number of purchases per customer for our top 10. 

from scipy.stats import invgamma
customer_scale = p*(v+returning_customers_summary['frequency']*returning_customers_summary['monetary_value'])
customer_shape = p*returning_customers_summary['frequency']+q
ez = customer_scale/(customer_shape-1)
plt.figure(figsize=(12,6))
plt.xlabel('$ Profit')
for ind,i in enumerate(best_projected_cust_mbg.index.tolist()):
    ccig = invgamma.rvs(customer_shape[i], scale=customer_scale[i], size = 10000)
    plt.text(ez[i]+5, invgamma.pdf(ez[i], customer_shape[i], scale=customer_scale[i]),
             '$'+str(round(ez[i],2))+'| '+str(int(returning_customers_summary.loc[i,'frequency'])))
    sns.distplot(ccig, hist=False, rug=False, label=str(i));

# 18308: corporate 
from lifetimes.plotting import plot_history_alive
import matplotlib.pyplot as plt
transaction_data = pd.read_csv('CLTV_BeforeLTV.csv')
fig = plt.figure(figsize=(12,8))
id = 18308
days_since_birth = 90
sp_trans = transaction_data.loc[transaction_data['user_id'] == id]
plot_history_alive(mbg, days_since_birth, sp_trans, 'purchase_date')

# 17499: personal 
fig = plt.figure(figsize=(12,8))
id = 17499
days_since_birth = 90
sp_trans = transaction_data.loc[transaction_data['user_id'] == id]
plot_history_alive(mbg, days_since_birth, sp_trans, 'purchase_date')

# merging average profit per customer to our original data set! 
merge = transaction_data.merge(data, left_on='user_id', right_on='user_id')
merge.to_csv("CLTVData_Monetary_Merged.csv", index=False, header=True)


# Since GGF model excluded onetime buyers, let's insert the profit per customer values for onetime buyers.
# also create customer type column to distinguish onetime vs repeat customers
customer_type = lambda row: 'onetime' if row.monetary_value == 0 else 'repeat'
onetimebuyers = lambda row: row.restaurant_total if row.monetary_value == 0 else row.monetary_value
merge['customer_type'] = merge.apply(customer_type, axis=1)
merge['monetary_value'] = merge.apply(onetimebuyers, axis=1)

merge['monetary_value'].describe()
# onetimebuyers no longer have 0 as their monetary value (average profit)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score

merge = pd.read_csv("CLTVData_Monetary_Merged.csv")
merge.set_index('user_id', inplace=True)

# drop phone orders to avoid multicollinearity 
# if source is call, shipping type & payment are phone
merge = merge[merge.shipping_type != "phone"]

# set X and Y 
X=merge[['source', 'shipping_type', 'payment_method', 'purchase_day', 'time_bracket']]
Y=merge['monetary_value']

# dummy variables 
X=pd.get_dummies(X)
# agg by sum, group by user id: only 7953 unique customers 
X= X.groupby(['user_id']).sum()
Y= Y.groupby(['user_id']).mean()

# add f & r to X 
rf=merge[['frequency', 'recency']]
rf=rf.groupby(['user_id']).mean()
result= pd.merge(X, rf, on='user_id')
result.info()
X = result
```
